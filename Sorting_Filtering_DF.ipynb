# Filtering DataFrames:
    By passing the logical test into the .loc[] accessor, just like we have done in Series or NumPy etc

```python
df.loc[df['date'] == '28-01-2016']
```

```python
df.loc[df['date'] == '05-02-2016', ['date', "sales"]].head()
```

```python
conds = df['family'].isin(['AUTOMOTIVE', 'Books']) & (df['sales'] > 0)
```

```python
df.loc[conds]
```

```python
df.query("family in ['AUTOMOTIVE', 'BABY CARE'] and sales > 0")
```

```python
avg_sales = df.loc[:, 'sales'].mean()
```

```python
avg_sales
```

```python
df.query("family in ['AUTOMATIVE', 'BABY CARE'] and sales > @avg_sales")
```

# Sorting DataFrames by Indices
   by using '.sort_index()' method
    
    sort rows (axis = 0) by default 
    
    specify, axis = 1 to sort the columns

```python
conditions = df.family.isin(["AUTOMOTIVE", "BABY CARE", "BEAUTY"])
sample_df = df[conditions].sample(5, random_state=2021)
sample_df
```

```python
sample_df.sort_index(ascending=False)
```

```python
conditions = df.family.isin(["AUTOMOTIVE", "BABY", "BEAUTY"])
sample_df = df[conditions].sample(5, random_state=2021)
sample_df
```

```python
sample_df.sort_index(axis = 1, inplace=True)
```

```python
sample_df
```

```python
sample_df
```

```python
sample_df.sort_values('store_nbr')
```

```python
sample_df.sort_values(['family', 'sales'], ascending=[True, False])
```
